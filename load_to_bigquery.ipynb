## 2. load job to bigquery

```python
# schema
schema = [
    bigquery.SchemaField("study_id", "STRING"),
    bigquery.SchemaField("patient_id", "STRING"),
    bigquery.SchemaField("sample_id", "STRING"),
    bigquery.SchemaField("age_at_diagnosis", "FLOAT"),
    bigquery.SchemaField("type_of_breast_surgery", "STRING"),
    bigquery.SchemaField("cancer_type", "STRING"),
    bigquery.SchemaField("cancer_type_detailed", "STRING"),
    bigquery.SchemaField("cellularity", "STRING"),
    bigquery.SchemaField("chemotherapy", "STRING"),
    bigquery.SchemaField("pam50_claudin_low_subtype", "STRING"),
    bigquery.SchemaField("cohort", "FLOAT"),
    bigquery.SchemaField("er_status_measured_by_ihc", "STRING"),
    bigquery.SchemaField("er_status", "STRING"),
    bigquery.SchemaField("neoplasm_histologic_grade", "FLOAT"),
    bigquery.SchemaField("her2_status_measured_by_snp6", "STRING"),
    bigquery.SchemaField("her2_status", "STRING"),
    bigquery.SchemaField("tumor_other_histologic_subtype", "STRING"),
    bigquery.SchemaField("hormone_therapy", "STRING"),
    bigquery.SchemaField("inferred_menopausal_state", "STRING"),
    bigquery.SchemaField("integrative_cluster", "STRING"),
    bigquery.SchemaField("primary_tumor_laterality", "STRING"),
    bigquery.SchemaField("lymph_nodes_examined_positive", "FLOAT"),
    bigquery.SchemaField("mutation_count", "FLOAT"),
    bigquery.SchemaField("nottingham_prognostic_index", "FLOAT"),
    bigquery.SchemaField("oncotree_code", "STRING"),
    bigquery.SchemaField("overall_survival_months", "FLOAT"),
    bigquery.SchemaField("overall_survival_status", "STRING"),
    bigquery.SchemaField("pr_status", "STRING"),
    bigquery.SchemaField("radio_therapy", "STRING"),
    bigquery.SchemaField("relapse_free_status_months", "FLOAT"),
    bigquery.SchemaField("relapse_free_status", "STRING"),
    bigquery.SchemaField("number_of_samples_per_patient", "INTEGER"),
    bigquery.SchemaField("sample_type", "STRING"),
    bigquery.SchemaField("sex", "STRING"),
    bigquery.SchemaField("3_gene_classifier_subtype", "STRING"),
    bigquery.SchemaField("tmb_nonsynonymous", "FLOAT"),
    bigquery.SchemaField("tumor_size", "FLOAT"),
    bigquery.SchemaField("tumor_stage", "FLOAT"),
    bigquery.SchemaField("patient_vital_status", "STRING"),
]
```

```python
# job configuration
job_config = bigquery.LoadJobConfig(
    schema=schema,
    write_disposition="WRITE_TRUNCATE",
)
```

```python
# table ref
dataset_id = 'breast_cancer'
table_id = 'breast_cancer'

table_id = f'{project_id}.{dataset_id}.{table_id}'
```

```python
# run job
job = client.load_table_from_dataframe(df,
                                       destination=table_id,
                                       job_config=job_config)
job.result()
```

```python
# check load job
table = client.get_table(table_id)  # Make an API request.
print(
    "Loaded {} rows and {} columns to {}".format(
        table.num_rows, len(table.schema), table_id
    )
)
```
